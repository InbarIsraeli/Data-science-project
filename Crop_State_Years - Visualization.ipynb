# Visualization
In this notebook we will visualize the effect of greenhouse gas affect on the crops along the years 1984-2019 in global growth, the amount of production is by tonnes.

1. compare between the crops/states to see who was the most\least affected by the gas than others along the years
    * which crop/state was the most prodacted althoug the gas levels went higher every year
    * which crop/state was the least prodacted because the gas levels went higher every year
    
    
2. compare between the amount of production for a specific crop/state along the years to see if it was affected by the gas

we can know all that from the gas levels and the amount of production

### gas levels in 2019 and 1991

```python
#in 2019 the gas levels where: 0.4105	332.0	1877
sum=0.4105+332+1877
x=0.4105/sum
y=332/sum
z=1877/sum
print(x,y,z)

plt.axis("equal")
lab=['CO2_levels', 'N2O_levels','CH4_levels']
ratios=[x,y,z]
plt.pie(ratios,data=df_Crops_Count,labels=lab,radius=2,autopct='%1.1f%%',explode = [0, 0, 0.10],startangle=(-180 * ratios[0]))
plt.show()
```

```python
#in 1991 the gas levels where: 0.3555	309.3	1735
sum=0.3555+309.3+1735
x=0.3555/sum
y=309.3/sum
z=1735/sum
print(x,y,z)

plt.axis("equal")
lab=['CO2_levels', 'N2O_levels','CH4_levels']
ratios=[x,y,z]
plt.pie(ratios,data=df_Crops_Count,labels=lab,radius=2,autopct='%1.1f%%',explode = [0, 0, 0.10],startangle=(-180 * ratios[0]))
plt.show()
```
