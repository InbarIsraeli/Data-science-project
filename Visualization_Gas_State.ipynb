### Bar plots of the state during the years 

```python
title="The effecr of Greenhous gas in Afghanistan"
ax1 = df_gas_State.groupby("year").sum()["Afghanistan"].plot.bar(title=title,color = 'pink',fontsize=18,figsize=(20,10))
ax1.title.set_size(25)
ax1.set_ylabel("Value in Tons",fontsize=18)
ax1.set_xlabel("year",fontsize=18)
```
