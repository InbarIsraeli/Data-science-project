# Visualization
In this notebook we will visualize the effect of greenhouse gas affect on the crops along the years 1984-2019 in global growth, the amount of production is by tonnes.

1. compare between the kinds to see who was the most\least affected by the gas than others along the years
    * which kind was the most prodacted althoug the gas levels went higher every year
    * which kind was the least prodacted because the gas levels went higher every year
    
    
2. compare between the amount of production for a specific kind along the years to see if it was affected by the gas

we can know all that from the gas levels and the amount of production

### Bar plots of the crops during the years 

```python
title="The production amount of Apples"
ax1 = df_Crops_Count.groupby("year").sum()["Apples Count"].plot.bar(title=title,color = '#D933FF',fontsize=18,figsize=(20,10))
ax1.title.set_size(25)
ax1.set_ylabel("Value in Tonnes",fontsize=18)
ax1.set_xlabel("year",fontsize=18)
```

```python
#in 2019 the gas levels where: 0.4105	332.0	1877
sum=0.4105+332+1877
x=0.4105/sum
y=332/sum
z=1877/sum
# print(x,y,z)

plt.axis("equal")
lab=['CO2_levels', 'N2O_levels','CH4_levels']
ratios=[x,y,z]
plt.pie(ratios,data=df_Crops_Count,labels=lab,radius=2,autopct='%1.1f%%',explode = [0, 0, 0.10],startangle=(-180 * ratios[0]))
plt.show()
```

```python
#in 1991 the gas levels where: 0.3555	309.3	1735
sum=0.3555+309.3+1735
x=0.3555/sum
y=309.3/sum
z=1735/sum
# print(x,y,z)

plt.axis("equal")
lab=['CO2_levels', 'N2O_levels','CH4_levels']
ratios=[x,y,z]
plt.pie(ratios,data=df_Crops_Count,labels=lab,radius=2,autopct='%1.1f%%',explode = [0, 0, 0.10],startangle=(-180 * ratios[0]))
plt.show()
```
