```python
import time 
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file_name = '.' + os.sep + 'data' + os.sep + 'FAOSTAT_data.csv'
df = pd.read_csv(file_name)
df
```

```python
df.describe(include='all')
```

```python
df.info()
```

*we found that in 'Flag' and 'Value' columns there are missing values*

```python
#Removing the 'Flag' column from the data set to get a cleaner data 
df_copy=df.copy()
df_copy=df_copy.drop(columns='Flag')
df_copy
```

```python
df_copy.info()
```

```python
#Removing the missing Values from the data set to get a cleaner data 
df_cleaned = df_copy.copy()

for row in df_cleaned :
    df_cleaned = df_cleaned.dropna( axis=0, how='any', subset=[row])
    
df_cleaned
```

```python
#checking the data for more missing valuse
df_cleaned.info()
```

*The data is clean and have more then 50,000 values. we will save it in a new csv file*

```python
df_cleaned.to_csv("data/FAOSTAT_cleaned_data.csv", index=False)
```
