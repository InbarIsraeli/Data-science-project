In this notebook we will split the Gas_Crops_data data frame, between the years *1984-2019*, to 6 diffrent:
* Crops - Apples, Potatoes, Cocoa, Vanila, Wheat, Banana
* States - Afghanistan, Israel, Zimbabwe, Colombia, Denmark, Mexico    

the amount of prodaction is by *tonnes*.

We will merge the amount of growth for each crop/state by year in a new data frame

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings("ignore")
import time 
import os
%matplotlib inline
```

```python
df_gas_crops = pd.read_csv('./data/Gas_Crops_data.csv')
df_gas_crops
```

```python
df_gas = pd.read_csv('./data/greenhouse gas.csv')
df_gas
```

### Spliting to Crops: 
Apples, Potatoes, Cocoa, Vanila, Wheat, Banana

```python
df_Crops_Count = df_gas.copy()

df_Crops_Count ['Apples Count'] = 0
df_Crops_Count ['Potatoes Count'] = 0
df_Crops_Count ['Cocoa Count'] = 0
df_Crops_Count ['Vanilla Count'] = 0
df_Crops_Count ['Wheat Count'] = 0
df_Crops_Count ['Bananas Count'] = 0

df_Crops_Count 
```

```python
df_Crops_Count.columns
```

```python

for i,valA in enumerate(df_gas_crops['Year']):
    for j, valB in enumerate(df_Crops_Count['year']):
        if (valA == valB):    
            if(df_gas_crops['Item'][i]=='Apples'):
                df_Crops_Count['Apples Count'][j] = df_Crops_Count['Apples Count'][j] + df_gas_crops['Value'][i]

            if(df_gas_crops['Item'][i]=='Potatoes'):
                df_Crops_Count['Potatoes Count'][j] = df_Crops_Count['Potatoes Count'][j] + df_gas_crops['Value'][i]
            
            if(df_gas_crops['Item'][i]=='Cocoa, beans'):
                df_Crops_Count['Cocoa Count'][j] = df_Crops_Count['Cocoa Count'][j] + df_gas_crops['Value'][i]
                
            if(df_gas_crops['Item'][i]=='Vanilla'):
                df_Crops_Count['Vanilla Count'][j] = df_Crops_Count['Vanilla Count'][j] + df_gas_crops['Value'][i]
            
            if(df_gas_crops['Item'][i]=='Wheat'):
                df_Crops_Count['Wheat Count'][j] = df_Crops_Count['Wheat Count'][j] + df_gas_crops['Value'][i]
            
            if(df_gas_crops['Item'][i]=='Bananas'):
                df_Crops_Count['Bananas Count'][j] = df_Crops_Count['Bananas Count'][j] + df_gas_crops['Value'][i]
```

```python
df_Crops_Count
```

```python
df_Crops_Count.to_csv("data/Crops_years.csv", index=False)
```
